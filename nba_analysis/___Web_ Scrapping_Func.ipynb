# Scraping data from http://mcubed.net website

```python
#importation of libraries and making connection to the website

from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
import pandas as pd
import numpy as np

myurl = 'http://mcubed.net/nba/fw2000.shtml'
uClient = uReq(myurl)
page_html = uClient.read()
uClient.close()
page_soup = soup(page_html, "html.parser")
```

```python
#overview of the website

page_soup
```

```python
#Groupping necessary portion of the website

grouped = page_soup.findAll('span')
```

```python
#First scrap team names and corresponding links, and save it in a list called, link_name

temp = []
link_name = []

for i in range(0,len(grouped)):
    temp.append(str(grouped[i]).split('"')[3])
    temp.append(str(grouped[i]).split('"')[4].split(">")[1].replace('</a', '').replace(' ','_'))

    link_name.append(temp)
    temp = []
```

```python
link_name
```

```python
#making functions for scraping all team records from website

def scrapping_func(link, name):
    teamurl = link
    uClient = uReq(teamurl)
    teampage_html = uClient.read()
    uClient.close()
    
    teampage_soup = soup(teampage_html, "html.parser")
    group_info = teampage_soup.findAll("span")
    
    team_year = []
    for i in range(0,len(group_info)):
        team_year.append(str(teampage_soup.findAll("span")[i]).split("  ")[0:3][0].split(">")[2].replace("</a",""))

    team_record_W = []
    team_record_L = []

    for i in range(0, len(group_info)):
        team_record_W.append(str(teampage_soup.findAll("span")[i]).split()[3])
        team_record_L.append(str(teampage_soup.findAll("span")[i]).split()[4])
        
    team_name = pd.DataFrame(team_record_W, team_year, columns=['W'])
    team_name['L'] = team_record_L
    
    export_csv = team_name.to_csv(name +'.csv')
    
    
def scrapping_all_func(mylist = []):
    for i in range(0,len(link_name)):
        link = mylist[i][0]
        name = mylist[i][1]
        scrapping_func(link, name)
```
